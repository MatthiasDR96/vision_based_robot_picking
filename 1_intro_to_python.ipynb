> Momenteel zit je in een speciaal Python script, een notebook genaamd. Dit zorgt ervoor dat je interactieve Python programma's kan maken en delen met de wereld. In onderstaand voorbeeld wordt 'Hello World' geprint in de console. Je kan de code cel uitvoeren door links op het play-symbool te drukken.



```python
# Hier wordt 'Hello World' uitgeprint
print('Hello World')
```



> In Python kan je met 'print' zo alles printen, van strings tot arrays. Je ziet dat je code kan uitcommenten met een '#'. In de volgende cel worden twee variabelen gecreëerd en opgeteld.




```python
a = 5
b = 2
c = a + b
print(c)
```


> Je kan zien dat een variabele niet gedeclareerd moet worden zoals in bvb Java. Je hoeft dus niet aan Python te zeggen wat er in 'a' zal komen.




```python
print(a)
a = 'Hello World'
print(a)
```



> Met bovenstaande cel kan je zien dat 'a' eerst een integer bevat en daarna een string. Python kan ook erg makkelijk met arrays werken. Een matrix van 3x3 maak je als volgt:



```python
array = [[1, 2, 3], [1, 2, 3], [1, 2, 3]]
print(array)
```



> Python kan bovenstaande som ook in een functie gieten. Dit gebeurt als volgt:



```python
def som(a, b):
  c = a + b
  return c
```


> Als je bovenstaande cel uitvoert dan zie je geen output, je moet hiervoor de functie eerst nog oproepen. Dit doe je zoals in volgende cel. Let op, onderstaande code zal een error genereren.




```python
c = som(a, b)
print(c)
```



> De code gaat in error omdat de variabelen 'a' en 'b' die je aan de functie geeft hierboven reeds gedefiniëerd waren (je hebt de cellen met toekenning van a en b uitgevoerd) en omdat 'a' nog een string bevat. En een string kan je niet optellen met een integer. Je moet dus eerst nog 'a' toekennen aan een integer vooraleer je de som kan maken.



```python
a = 5
c = som(a, b)
print(c)
```



> Deze code geeft terug een goed resultaat.
